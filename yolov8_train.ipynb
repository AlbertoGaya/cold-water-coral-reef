```python
import os

# Ruta de la carpeta que contiene las imÃ¡genes
carpeta_imagenes = "IC222_TF18"

# Obtener una lista de nombres de archivos en la carpeta
nombres_archivos = os.listdir(carpeta_imagenes)

# Ruta para el archivo de texto donde se escribirÃ¡n los nombres
archivo_txt = "nombres_imagenes.txt"

# Escribir los nombres de los archivos en el archivo de texto
with open(archivo_txt, 'w') as archivo:
    for nombre in nombres_archivos:
        archivo.write(nombre + '\n')

print("Se ha creado el archivo", archivo_txt)
```

```python
import os
import re
import shutil

def copiar_imagenes_impares(origen, destino):
    # Obtener la lista de archivos en la carpeta de origen
    archivos = os.listdir(origen)
    
    # Crear la carpeta de destino con el nombre original mÃ¡s '_pair'
    carpeta_destino = destino + "_pair"
    os.makedirs(carpeta_destino, exist_ok=True)

    # Recorrer todos los archivos en la carpeta de origen
    for archivo in archivos:
        # Verificar si el archivo es una imagen JPG
        if archivo.lower().endswith(".jpg"):
            # Utilizar expresiÃ³n regular para extraer el nÃºmero del nombre del archivo
            match = re.search(r'A0410_TF09_(\d+)\.JPG', archivo)
            if match:
                numero_archivo = int(match.group(1))
                
                # Verificar si el nÃºmero de archivo es impar
                if numero_archivo % 2 != 0:
                    # Copiar el archivo a la carpeta de destino
                    shutil.copy(os.path.join(origen, archivo), carpeta_destino)
                    print(f"Imagen impar {archivo} copiada a {carpeta_destino}")

# Ejemplo de uso
carpeta_origen = "A0410_TF09"
carpeta_destino = os.path.basename(carpeta_origen)  # Usar el mismo nombre que la carpeta original

copiar_imagenes_impares(carpeta_origen, carpeta_destino)
```
